```python
import os
import pandas as pd

# Clone the repository
os.system("git clone https://github.com/gauthamp10/Google-Playstore-Dataset.git")

# Navigate to the dataset directory
os.chdir("Google-Playstore-Dataset/dataset/")

# Extract all tar.gz files in the directory
os.system("for f in *.tar.gz; do tar -xvf \"$f\"; done")

# Merge all Part files into a single CSV file
os.system("cat Part?.csv > Google-Playstore-Dataset.csv")

# Read the CSV file as a pandas dataframe
df = pd.read_csv("Google-Playstore-Dataset.csv")

# Print the number of rows and columns in the dataframe
print("Number of rows:", len(df))
print("Number of columns:", len(df.columns))
```
